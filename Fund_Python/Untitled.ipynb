```python
from pydantic import BaseModel
from typing import Optional
class User(BaseModel):
	user_id: int
	name: str
	age: int
	address: Optional[str]
```

```python
jaja = User(user_id= 1,name='King',age='30',address= '!!2323')
jaja.address
```
