```python
d = {}
with open('scores.csv','r') as f:
    for line in f:
        name,_,score = line.split(',')
        score = int(score.strip())
        if name in d:
            d[name].append(score)
        else:
            d[name] = [score]
for i in d:
    print(f'Name {i} Max score: {max(d[i])} Min score: {min(d[i])} Avg score: {(sum(d[i])/len(d[i]))}')
```
