#### Common Status Codes

1.  200: OK (The request was successful).
2.  201: Created (The resource was successful created).
3.  400: Bad Request (The server could not understand the request)
4.  401: Unauthorized (Authentication is required).
5.  404: Not Found (The requested resource could not be found).
6.  500: Internal Server Error (The server encountered an error).

```python
# URL of the API endpoint
url = 'https://jsonplaceholder.typicode.com/posts/1'

# Sending a GET request
response  = requests.get(url)

# Handling different status codes
if response.status_code == 200:
	print('Request was succesful')
	print(response.json())
elif response.status_code == 400:
	print('Cant request')
else:
	print(f'Failed with status code: {response.status_code}')
```

# Cryto currency exchange APIs

```python
import requests

url = "https://api.coingecko.com/api/v3/simple/price"

para = {
    'ids':'bitcoin,ethereum', # Assign id to string
	'vs_currencies':'USD'
}

response = requests.get(url, params=para) # pass parameter for api which use

data = response.json()
```

```python
for key in data:
	print(f'In {key} have {data[key]}')
```
