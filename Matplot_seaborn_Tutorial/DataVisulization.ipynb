```python
import pandas as pd
```

```python
df = pd.read_excel('linechart.xlsx')
df
```

```python
from matplotlib import pyplot as  plt
```

```python
plt.figure(figsize=(12,4))

plt.plot(df['Quarter'], df['Fridge'], label='Fridge', color='skyblue')
plt.plot(df['Quarter'], df['Dishwasher'], label='Dishwasher', color='green')
plt.plot(df['Quarter'], df['Washing Machine'], label='Washing Machine', color='orange')

plt.title('Product Sales')
plt.ylabel('Revenue (ml $)')
plt.xlabel('Financial quarter')

plt.legend()
plt.show()
```

```python
total_revenue = df[['Dishwasher','Fridge','Washing Machine']].sum()
total_revenue
```

```python
plt.pie(total_revenue, labels=total_revenue.index, autopct='%1.2f%%', explode=[0,0.1,0],
	   shadow=True,startangle=-90)

plt.show()
```

```python
df.plot(kind='bar',x='Quarter')

plt.xticks(rotation=45)

plt.show()
```
