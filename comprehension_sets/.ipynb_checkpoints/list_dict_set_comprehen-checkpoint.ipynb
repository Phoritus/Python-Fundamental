```python
number = [1,2,3,4,5]
sqr_number = [] 

for n in number:
	sqr_number.append(n*n)
sqr_number  # Use for loop
```

```python
sqr_number = [i*i for i in number]
sqr_number # Use list comprehension
```

```python
import time

long_list = range(100000)
sqr_number
start = time.time()
for i in long_list:
	sqr_number.append(i*i)
end = time.time()
print(end-start)
```

```python
start = time.time()
sqr_number = [i*i for i in long_list]
end = time.time()
print(end-start)
```

```python
price = [100,120,80,50]
earning_per_share = [9,15,40,2]
```

```python
pe = [round((p/e),2) for p,e in zip(price,earning_per_share)]
pe 
```

```python
price = [100,120,80,50]
earning_per_share = [0,15,40,2]
```

```python
pe = [round((p/e),2) if e != 0 else -1 for p,e in zip(price,earning_per_share)]
pe
```

```python
word = ['apple','banana','cherry','date']
w_lenght = [len(s) for s in word]
w_lenght
```

```python
prices = [120,99,150,80,200]
dis_10 = [(i-(i*0.1)) if i > 100 else i for i in prices] # Use condition in list comprehension
dis_10
```

```python
names = ['Alice','Bob','Charlie']
ages = [25,30,35]
```

```python
n_age = {names[i]:ages[i] for i in range(len(names))} # Dictionary comprehension
n_age
```

```python
transactions = [
	{'cutomer_name':'Mira', 'amount':56, 'date':'2024-08-01'},
	{'cutomer_name':'Mira', 'amount':12.3, 'date':'2024-09-05'},
	{'cutomer_name':'Mohan', 'amount':21, 'date':'2024-09-03'},
	{'cutomer_name':'Mohan', 'amount':36.8, 'date':'2024-09-10'},
	{'cutomer_name':'Lucy', 'amount':66.3, 'date':'2024-09-10'}
]
```

```python
customer = {name['cutomer_name'] for name in transactions} # Set comprehension
customer
```
