```python
transactions = [
	{'cutomer_name':'Mira', 'amount':56, 'date':'2024-08-01'},
	{'cutomer_name':'Mira', 'amount':12.3, 'date':'2024-09-05'},
	{'cutomer_name':'Mohan', 'amount':21, 'date':'2024-09-03'},
	{'cutomer_name':'Mohan', 'amount':36.8, 'date':'2024-09-10'},
	{'cutomer_name':'Lucy', 'amount':66.3, 'date':'2024-09-10'}
]
```

```python
customer = set()
for name in transactions:
	customer.add(name['cutomer_name'])
customer
```

```python
customer_list = list(customer)
customer_list
```
