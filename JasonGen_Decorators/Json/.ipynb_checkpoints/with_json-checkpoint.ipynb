```python
# JSON string
json_string = '''
{
    "name": "Maulik",
    "age": 30,
    "is_employee": true,
    "skills": ["Python", "Data Analysis", "Machine Learning"],
    "address": {
        "street": "123 Main St",
        "city": "Wonderland"
    }
}
'''
type(json_string)
```

```python
import json

j_f = json.loads(json_string)
j_f # Make dictionary
```

```python
type(j_f)
```

```python
j_f['name'],j_f['address'] # Access value in dict
```

```python
data = {
    "name": "Bob",
    "age": 25,
    "is_employee": False,
    "skills": ["JavaScript", "Web Development"],
    "address": {
        "street": "456 Elm St",
        "city": "Springfield"
    }
}
```

```python
data_str = json.dumps(data)
data_str # Use dumps for convert json file to sring
```

```python
type(data_str)
```

# Write Json data to file

```python
with open('output.json','w') as f:
	json.dump(data,f, indent=4)
	
```

```python
json_string = '''
{
    "books": [
        {"title": "The Magic Of Thinking Big", "author": "David J. Schwartz", "published": 1959},
        {"title": "Chanakya Neeti", "author": "B.K. Chaturvedi", "published": 2009},
        {"title": "How to Win Friends and Influence People", "author": "Dale Karnegie", "published": 1936}
    ],
    "status_code": 200
}
'''
```

```python
data = json.loads(json_string)
data # Convert Json file to dictionary
```

```python
books = data['books']
books # Cleaning dict list
```

```python
book_list = [ {'title':book['title'], 'author':book['author']} for book in books]
book_list # Assign list to keep title,author
```

```python
book_list[0]
```

```python
book_list[0]['author']
```

# Error Handling with Json

```python
# Malformed JSON string
json_string = '''
{
    "name": "Krishna",
    "age": 30
    "is_employee": true
}
'''
```

```python
json.loads(json_string) # Don't use try except
```

```python
try:
	json.loads(json_string)
except Exception as e:
	print('Falie to parse Json',e) # Use try except
```
