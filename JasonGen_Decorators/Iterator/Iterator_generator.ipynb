```python
class RemoteControl:
	def __init__(self):
		self.channal = ['CNN','TNBC','BI']
		self.index = -1

	def __next__(self):
		self.index += 1
		if self.index == len(self.channal):
			raise StopIteration
		else:
			return self.channal[self.index]

	def __iter__(self):
		return self # Now create function to structure of iter
```

```python
ch = RemoteControl()
for cha in ch:
	print(cha)
```

```python
def remote_control():
	yield 'CNN'
	yield 'TNBC'
	yield 'TNN' # Now the best way is use Generators the logic is same with iter but shorter
```

```python
rc = remote_control()
type(rc)
```

```python
for c in rc:
	print(c)
```

```python
def remote_control():
	for i in range(1,4):
		with open(f'stock_{i}.txt','r') as f:
			yield f.read() # Use yield to make function become generator
```

```python
df = []
for data in remote_control():
	df.append(data)
for d in df:
	print(d)
```

```python
def fibo_gen():
	a,b = 0,1
	while True:
		yield a
		a, b = b , a+b
# This fibonacci by use generator
```

```python
n = fibo_gen()
for i in range(10):
	print(next(n))
```
