```python
import time
```

```python
def get_square(numbers):
	start = time.time()
	result = []
	for n in numbers:
		result.append(n*n)
	end = time.time()
	print(f'get_square took {(end - start)*1000} mil seconds')

	return result
```

```python
def get_cube(numbers):
	start = time.time()
	result = []
	for n in numbers:
		result.append(n*n*n)
	end = time.time()
	print(f'get_square took {(end - start)*1000} mil seconds')
	return result
```

```python
_ = get_square(range(1000000))
```

```python
_ = get_cube(range(1000000))
```

```python
def timer(func): # Create decorator timer function
	def wrapper(*args,**kwargs):
		start = time.time()
		result = func(*args,**kwargs)
		end = time.time()
		print(f'Functino {func.__name__} take {(end - start)*1000} mil seconds')
		return result
	return wrapper
```

```python
@timer # Use decorator timer function
def get_cube(numbers):
	result = []
	for n in numbers:
		result.append(n*n*n)
	return result
```

```python
_ = get_cube(range(1000000))
```

```python
@timer
def get_sqaure(numbers):
	result = []
	for n in numbers:
		result.append(n*n)
	return result
```

```python
_ = get_sqaure(range(1000000))
```
