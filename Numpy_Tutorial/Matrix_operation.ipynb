```python
import numpy as np
```

```python
q1 = np.array([
    [200,220,250],
    [150,180,210],
	[300,330,360]
])

q2 = np.array([
	[209,231,259],
	[155,192,222],
	[310,340,375]
])
```

```python
q1 + q2
```

```python
q2 - q1
```

```python
(q2-q1)*100/q1
```

```python
prices = np.array([
	[10,12,11],
	[8,9,10],
	[15,16,17]
])

q1_revenue = q1 * prices
q1_revenue
```

```python
q1_discount = q1_revenue*0.3
q1_discount
```

```python
net_worth = q1_revenue - q1_discount
net_worth
```
