```python
import pandas as pd
```

```python
df = pd.read_csv('weather_data.csv',parse_dates=['day'],index_col=['day'])
df
```

```python
df.fillna(0,inplace=True)
```

```python
df
```

```python
df.fillna({
	'temperature': df.temperature.mean(),
	'windspeed': df.windspeed.mean(),
	'event': 'No Event'
})
```

```python
df
```

```python
df.fillna(method='ffill') # Forward fill
```

```python
df.fillna(method='bfill') # Backward fill
```

```python
df.fillna(method='bfill',axis='columns')
```

```python
df.fillna(method='ffill',limit=1) # Limit forward to fill
```

```python
df.interpolate() # Find median between
```

```python
df.dropna() # Drop if null.
```

```python
df.dropna(how='all') # All nulls dropped.
```

```python
df.dropna(thresh=1) # Same with how='all'
```

```python
df.dropna(thresh=2) # If there are 2 nulls, you can drop it.
```
