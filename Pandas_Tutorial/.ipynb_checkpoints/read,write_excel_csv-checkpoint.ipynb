```python
df_movie = pd.read_excel('movies_db.xlsx','movies',index_col='movie_id')
df_movie.head() # Read Excel file
```

```python
df_financial = pd.read_excel('movies_db.xlsx','financials',index_col='movie_id')
df_financial.head() 
```

```python
def convert_d(d):
	if d == '$$' or d == 'Dollars':
		return 'USD'
	else:
		return d # Convert '$$' and 'Dollars' to 'USD'

df_financial = pd.read_excel('movies_db.xlsx','financials',index_col='movie_id',converters={
	'currency': convert_d # Use with function
})
df_financial.head()
```

```python
df_merged = pd.merge(df_movie,df_financial,on='movie_id')
df_merged.head()
```

```python
df_merged.to_excel('movie_merged.xlsx') # Import excel file
```
