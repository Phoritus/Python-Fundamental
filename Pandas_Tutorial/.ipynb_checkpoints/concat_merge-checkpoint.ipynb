```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
	'city':['mumbai','delhi','banglore'],
	'temperature':[32,45,30],
	'humidity':[80,60,78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
	'city':['new york','chicago','orlando'],
	'temperature':[21,14,35],
	'humidity':[68,65,75]
})
us_weather
```

```python
df = pd.concat([india_weather,us_weather],ignore_index=True)
df # Concat data with seqeuntial number
```

```python
df = pd.concat([india_weather,us_weather], keys=['India','us'])
df # Use key like a hashmap
```

```python
df.loc['India'] # Locate value by use key
```

```python
temperature_df = pd.DataFrame({
	'city': ['mumbai','delhi','banglore'],
	'temperature': [32,45,30]
})
temperature_df
```

```python
windspeed_df = pd.DataFrame({
	'city': ['mumbai','delhi'],
	'windspeeds': [12,7]
},index=[0,1])
windspeed_df
```

```python
pd.concat([temperature_df,windspeed_df],axis=1) # Concat with column by use axis=1
```

```python
df1 = pd.DataFrame({
	'city': ['new york','chicago','orlando'],
	'temperature': [21,14,35]
})
df1
```

```python
df2 = pd.DataFrame({
	'city': ['chicago','new york','orlando'],
	'humidity': [65,68,75]
})
df2
```

```python
df3 = pd.merge(df1, df2, on='city')
df3 # Same with inner join
```

```python
df1 = pd.DataFrame({
	'city': ['new york','chicago','orlando','baltimore'],
	'temperature': [21,14,35,38]
})
df1 
```

```python
df2 = pd.DataFrame({
	'city': ['chicago','new york','san diego'],
	'humidity': [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1, df2, on='city',how='inner')
df3 # Inner join
```

```python
df3 = pd.merge(df1, df2, on='city', how='left')
df3 # Left join
```

```python
df3 = pd.merge(df1, df2, on='city', how='right')
df3 # Right join
```

```python
df3 = pd.merge(df1, df2, on='city',how='outer')
df3 # Outer join
```
